```python
import pandas as pd

df = pd.read_csv("ds_salaries.csv")

df.head()
```

```python
print(df.shape)
```

```python
df.info()
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df.head()
```

```python
experience_map = {
    'EN': 'Entry Level',
    'MI': 'Mid Level',
    'SE': 'Senior Level',
    'EX': 'Executive Level'
}

df['experience_level'] = df['experience_level'].map(experience_map)
```

```python
employment_map = {
    'FT': 'Full Time',
    'PT': 'Part Time',
    'CT': 'Contract',
    'FL': 'Freelance'
}

df['employment_type'] = df['employment_type'].map(employment_map)
```

```python
remote_map = {
    0: 'Onsite',
    50: 'Hybrid',
    100: 'Remote'
}

df['work_mode'] = df['remote_ratio'].map(remote_map)
```
